```python
import pandas as pd
```

```python
df = pd.read_csv("3_emocare_with_persona.csv")
```

```python
instruction = (
    "If you are a counsellor, please answer the questions based on the description of the patient."
)
```

```python
def create_prompt(row):
    return f"""{instruction}

Patient Persona:
{row['persona']}

Cultural Background:
{row['cultural_background']}

Patient Description:
{row['input']}

Response:"""
```

```python
df["prompt"] = df.apply(create_prompt, axis=1)
```

```python
final_df = df[["input", "output", "prompt"]]
```
