```python
import pandas as pd
```

```python
persona_df = pd.read_csv("1_en_IN-00000-of-00011.csv")
emocare_df = pd.read_csv("2_EmoCare.csv")
```

```python
persona_sample = persona_df.iloc[:len(emocare_df)].reset_index(drop=True)
```

```python
emocare_df = emocare_df.reset_index(drop=True)
```

```python
merged_df = pd.DataFrame({
    "persona": persona_sample["persona"],
    "cultural_background": persona_sample["cultural_background"],
    "input": emocare_df["input"],
    "output": emocare_df["output"]
})
```
