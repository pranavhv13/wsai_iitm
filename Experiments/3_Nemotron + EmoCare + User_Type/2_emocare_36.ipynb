## How the 36 Questions Were Selected

1. **Loaded data** - all 8,187 EmoCare Q&A rows and all 36 personas.
2. **Built a profile for each persona** - combined their `persona` summary, career goals, hobbies, marital status, sex, and age-stage into one descriptive text block (each weighted so it reads like a counselling narrative).
3. **Scored similarity** - used TF-IDF + cosine similarity to compare each persona's profile against every EmoCare `input` (the patient's own words).
4. **Filtered for length** - kept only EmoCare questions with 45+ words (at/above the corpus median), so answers are substantial but not sacrificing match quality - a pool of ~3,690 candidates.
5. **Optimized 1-to-1 assignment** - used the Hungarian algorithm to assign each persona its best possible match while guaranteeing all 36 questions are unique (no repeats across personas).
6. **Verified integrity** - confirmed every `instruction`, `input`, and `output` was copied exactly from EmoCare.csv, byte-for-byte, with no edits.

**Result:** 36 unique, moderately long (45-79 word), thematically-matched questions - one per persona.
