```python
import pandas as pd

df = pd.read_csv("1_personas_36.csv")

persona_columns = [
    "persona",
    "professional_persona",
    "linguistic_persona",
    "sports_persona",
    "arts_persona",
    "travel_persona",
    "culinary_persona",
    "cultural_background",
    "linguistic_background",
    "skills_and_expertise",
    "hobbies_and_interests",
    "career_goals_and_ambitions",
    "sex",
    "age",
    "marital_status",
    "education_level",
    "education_degree",
    "occupation",
    "zone",
    "state",
    "district"
]


def merge_persona(row):
    text = []

    for col in persona_columns:
        value = row[col]

        if pd.isna(value):
            continue

        label = col.replace("_", " ").title()

        text.append(f"{label}: {value}")

    return "\n\n".join(text)


new_df = pd.DataFrame()

new_df["persona_id"] = range(1, len(df) + 1)

new_df["persona"] = df.apply(merge_persona, axis=1)

new_df.to_csv("3_personas_merged.csv", index=False)

print("Created personas_merged.csv")
display(new_df.head())
```
